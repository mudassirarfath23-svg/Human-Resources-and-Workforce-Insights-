```python
# ================================
# HUMAN RESOURCE DATA ANALYSIS
# BASIC CLEANING + FEATURE ENGINEERING
# ================================

import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("E:\AAA_projects\HR\existing.csv")

# ================================
# 1. CHECK DATATYPES
# ================================
print(df.dtypes)
```

```python
# 2. CONVERT COLUMN NAMES TO LOWERCASE + UNDERSCORE FORMAT
# ================================
df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")
```

```python
df.columns
```

```python
# 3. CHECK NULL VALUES
# ================================
print(df.isnull().sum())
```

```python
# 4. REMOVE DUPLICATES
# ================================
df.drop_duplicates(inplace=True)
```

```python
# ================================
# 5. FEATURE ENGINEERING - NEW COLUMNS
# ================================

# Experience category
df['experience_category'] = pd.cut(
    df['totalworkingyears'],
    bins=[-1, 2, 5, 10, 20, 40],
    labels=['Fresher', 'Junior', 'Mid-Level', 'Senior', 'Expert']
)

# Income yearly
df['annual_income'] = df['monthlyincome'] * 12

# Salary efficiency
df['income_per_year_experience'] = df['monthlyincome'] / (df['totalworkingyears'] + 1)

# Distance category
df['distance_category'] = pd.cut(
    df['distancefromhome'],
    bins=[0, 5, 15, 30],
    labels=['Near', 'Moderate', 'Far']
)

# Satisfaction score combined
df['overall_satisfaction_score'] = (
    df['environmentsatisfaction'] +
    df['jobsatisfaction'] +
    df['relationshipsatisfaction']
)

# Promotion readiness indicator
df['promotion_ready'] = np.where(
    (df['performancerating'] >= 3) &
    (df['joblevel'] >= 2),
    'Yes',
    'No'
)

# Attrition risk score
df['attrition_risk_score'] = (
    df['overtime'].map({'Yes': 2, 'No': 0}) +
    (4 - df['jobsatisfaction']) +
    (4 - df['environmentsatisfaction'])
)

# Work stability score
df['work_stability_score'] = (
    df['totalworkingyears'] -
    df['numcompaniesworked']
)
```

```python
df.columns
```

```python
# Analysis Starts 
#Promotion Ready Employees by Department
df.groupby('department')['promotion_ready'].value_counts()
```

```python
#Work Stability by Age Group
df.groupby('agegroup')['work_stability_score'].mean()
```

```python
#High Attrition Risk Employees
df[df['attrition_risk_score'] >= 5]
```

```python
# 7. SAVE CLEANED FILE
# ================================
df.to_csv(r"E:\AAA_projects\HR\proposed.csv", index=False)
```
